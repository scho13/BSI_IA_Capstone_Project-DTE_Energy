```python
import pandas as pd
```

```python
electricity_data = pd.read_csv('AMI_Daily_Load_Res_zip.csv')
weather_data = pd.read_csv('Actual+Forecast Weather_Daily.csv')
```

```python
electricity_data['READ_DATE'] = pd.to_datetime(electricity_data['READ_DATE'])
weather_data['DATE_PART'] = pd.to_datetime(weather_data['DATE_PART'])
```

```python
# Convert DATE_PART in electricity_df to datetime
electricity_data['READ_DATE'] = pd.to_datetime(electricity_data['READ_DATE'])

# Convert DATE_PART in weather_df to datetime
weather_data['DATE_PART'] = pd.to_datetime(weather_data['DATE_PART'])

# Convert ZIP_CODE in electricity_df to string
electricity_data['ZIP_CODE'] = electricity_data['ZIP_CODE'].astype(str)

# Merge the datasets based on DATE_PART
merged_df = pd.merge(electricity_data, weather_data, left_on='READ_DATE', right_on='DATE_PART', how='inner')

merged_df = merged_df.drop(columns=['DATE_PART'])
merged_df = merged_df.dropna()

#Filter out with just the relevant stations
relevant_stations = ['KDTW', 'KFNT', 'KVLL']
filtered_df = merged_df[merged_df['STATIONID'].isin(relevant_stations)]

grouped = filtered_df.groupby('ZIP_CODE')
```

```python
# Find the correlations of all weather stations between DAY_KWH and AVG_TEMP for each zipcode and station
correlations = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_TEMP']

correlations2 = correlations.reset_index()
correlations_table = correlations2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_TEMP')

# Find the correlations of all weather stations between DAY_KWH and MAX_TEMP for each zipcode and station
correlations = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['MAX_TEMP']

correlations2 = correlations.reset_index()
correlations_table = correlations2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='MAX_TEMP')

# Find the correlations of all weather stations between DAY_KWH and AVG_DEWPT for each zipcode and station
correlations = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_DEWPT']

correlations2 = correlations.reset_index()
correlations_table = correlations2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_DEWPT')

# Find the correlations of all weather stations between DAY_KWH and AVG_RELHUM for each zipcode and station
correlations = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_RELHUM']

correlations2 = correlations.reset_index()
correlations_table = correlations2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_RELHUM')

# Find the correlations of all weather stations between DAY_KWH and AVG_WSPD for each zipcode and station
correlations = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_WSPD']

correlations2 = correlations.reset_index()
correlations_table = correlations2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_WSPD')

# Find the correlations of all weather stations between DAY_KWH and AVG_CLOUDCOVER for each zipcode and station
correlations = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_CLOUDCOVER']

correlations2 = correlations.reset_index()
correlations_table = correlations2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_CLOUDCOVER')
```

```python
# Make a table of the aggregated correlation coefficients for all the weather stations
correlations_avg_temp = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_TEMP']
correlations_max_temp = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['MAX_TEMP']
correlations_avg_dewpt = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_DEWPT']
correlations_avg_relhum = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_RELHUM']
correlations_avg_wspd = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_WSPD']
correlations_avg_cloudcover = merged_df.groupby(['ZIP_CODE', 'STATIONID']).corr().unstack()['DAY_KWH']['AVG_CLOUDCOVER']


correlations_avg_temp_2 = correlations_avg_temp.reset_index()
correlations_max_temp_2 = correlations_max_temp.reset_index()
correlations_avg_dewpt_2 = correlations_avg_dewpt.reset_index()
correlations_avg_relhum_2 = correlations_avg_relhum.reset_index()
correlations_avg_wspd_2 = correlations_avg_wspd.reset_index()
correlations_avg_cloudcover_2 = correlations_avg_cloudcover.reset_index()


correlations_table_avg_temp = correlations_avg_temp_2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_TEMP')
correlations_table_max_temp = correlations_max_temp_2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='MAX_TEMP')
correlations_table_avg_dewpt = correlations_avg_dewpt_2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_DEWPT')
correlations_table_avg_relhum = correlations_avg_relhum_2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_RELHUM')
correlations_table_avg_wspd = correlations_avg_wspd_2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_WSPD')
correlations_table_avg_cloudcover = correlations_avg_cloudcover_2.pivot_table(index='ZIP_CODE', columns='STATIONID', values='AVG_CLOUDCOVER')


correlations_table = correlations_table_avg_temp + correlations_table_max_temp + correlations_table_avg_dewpt + correlations_table_avg_relhum + correlations_table_avg_wspd + correlations_table_avg_cloudcover
correlations_table
```

```python
# Initialize an empty list to store results
avg_temp_results = []

# Iterate over each group
for zipcode, group in grouped:
    # Calculate correlation between 'DAY_KWH' and 'AVG_TEMP' for each station
    avg_temp_correlation = group.groupby('STATIONID').corr().unstack()['DAY_KWH']['AVG_TEMP']
    
    # Find the station with the highest correlation
    max_avg_temp_station = avg_temp_correlation.idxmax()
    
    # Get the highest correlation value
    max_avg_temp_value = avg_temp_correlation.max()
    
    # Append results to the list
    avg_temp_results.append({'ZIP_CODE': zipcode, 'AVG_TEMP': max_avg_temp_station})

# Create DataFrame from results
avg_temp = pd.DataFrame(avg_temp_results)
avg_temp
```
