```python
import pandas as pd
import numpy as np

import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
```

```python
SummerPeakHourlyUsage = pd.read_excel("~/Desktop/DTE Files/ServiceAreaHourlyLoad.xlsx")
```

```python
SummerPeakHourlyUsage.head()
```

```python
SummerPeakHourlyUsage['Date'] = pd.to_datetime(SummerPeakHourlyUsage['Date'])
SummerPeakHourlyUsage = SummerPeakHourlyUsage.dropna()
```

```python
KDTWHourly = pd.read_csv("~/Desktop/DTE Files/Actual+Forecast Weather_KDTW_Hourly.csv")
```

```python
KDTWHourly.head()
```

```python
KDTWHourly['VALIDDATETIME'] = pd.to_datetime(KDTWHourly['VALIDDATETIME'])
```

```python
KDTWHourly.rename(columns = {'VALIDDATETIME':'Date'}, inplace = True)
KDTWHourly['Date'] = KDTWHourly['Date'].dt.tz_localize(None)
```

```python
merged_df = pd.merge(SummerPeakHourlyUsage, KDTWHourly, on='Date', how='inner')
merged_df.drop(columns=['STATIONID'], inplace=True)
merged_df['High Usage'] = np.where(merged_df['GenOps Service Area'] >= 9000, 1, 0)
merged_df
```

```python
df = merged_df[merged_df['GenOps Service Area'] >= 9000]
df.drop(columns='WEATHER', inplace=True)
df
```

```python
grouped_df = df.groupby('Hr').size().reset_index(name='Count')
grouped_df
```

```python
#assign the cases with >9000 a value of 1, the others with a value of 0
#then create a model that works to predict whether or not an hour will have a usage of greater than 9000 based on the available characteristics
```

```python
model = smf.ols(" Q('GenOps Service Area') ~ Q('AVG_TEMP') ", data=merged_df).fit()
model.summary()
```

```python
model = smf.ols(" Q('GenOps Service Area') ~ Q('AVG_DEWPT') ", data=merged_df).fit()
model.summary()
```

```python
model = smf.ols(" Q('GenOps Service Area') ~ Q('AVG_RELHUM') ", data=merged_df).fit()
model.summary()
```

```python
model = smf.ols(" Q('GenOps Service Area') ~ Q('AVG_WSPD') ", data=merged_df).fit()
model.summary()
```

```python
model = smf.ols(" Q('GenOps Service Area') ~ Q('AVG_CLOUDCOVER') ", data=merged_df).fit()
model.summary()
```

```python
model = smf.ols(" Q('GenOps Service Area') ~ Q('AVG_TEMP') + Q('AVG_DEWPT') + Q('AVG_RELHUM') + Q('AVG_WSPD') + Q('AVG_CLOUDCOVER') ", data=merged_df).fit()
model.summary()
```
